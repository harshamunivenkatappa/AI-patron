```python
       
# pygame.init()
def passball(object1,target):
    print("passing the ball to","(",target.x,",",target.y,")")
    while abs(target.x - object1.x)>3 or abs(target.y - object1.y)>3:
        if (target.x-object1.x)<0:
            object1.x += -2.5
            object1.x += (target.x-object1.x)/1000000
        else:
            object1.x += 2.5
            object1.x += (target.x-object1.x)/1000000
        if (target.y-object1.y)<0:
            object1.y += -1
            object1.y += (target.y - object1.y)/1000000
        else:
            object1.y += 1
            object1.y += (target.y - object1.y)/1000000
        display()
        gameDisplay.blit(object1.img,(object1.x,object1.y))
        pygame.display.update()
    display()
    gameDisplay.blit(object1.img,(object1.x,object1.y))
    pygame.display.update()
    ball.x=object1.x
    ball.y=object1.y
    update_xy()
    time.sleep(2)
    clock.tick(30)
        
def dribble(object1):
    temp.x=(ball.x-object1.x)/2
    temp.y=(ball.y-object1.y)/2
    
    for i in range(10):
#         object1.x += 0.3
        object1.x += (temp.x-object1.x)/1000000
#         object1.y += 0.3
        object1.y += (temp.y-object1.y)/1000000
        gameDisplay.blit(ball.img,(object1.x,object1.y))
        pygame.display.update()
        time.sleep(.01)
        gameDisplay.blit(object1.img,(object1.x,object1.y))
        pygame.display.update()
        time.sleep(.1)
    ball.x=object1.x
    ball.y=object1.y
    update_xy()
    gameDisplay.blit(ball.img,(object1.x,object1.y))
    pygame.display.update()
    display()
    gameDisplay.blit(object1.img,(object1.x,object1.y))
    pygame.display.update()
    time.sleep(2)
    clock.tick(30)
    
def run(object1,target):
    temp.x=(target.x-object1.x)/5
    temp.y=(target.y-object1.y)/5
    
    for i in range(100):
#         object1.x += 0.3
        object1.x += (temp.x-object1.x)/1000000
#         object1.y += 0.3
        object1.y += (temp.y-object1.y)/1000000
#         gameDisplay.blit(ball.img,(object1.x,object1.y))
#         pygame.display.update()
#         time.sleep(1)
        gameDisplay.blit(object1.img,(object1.x,object1.y))
        pygame.display.update()
        time.sleep(.001)
    object1.x,object1.y=target.x,target.y
    display()
    update_xy()
    gameDisplay.blit(object1.img,(object1.x,object1.y))
    print(object1.name,"is running to","(",object1.x,",",object1.y,")")
    pygame.display.update()
    ball.x=object1.x
    ball.y=object1.y
    update_xy()
#     time.sleep(2)
    clock.tick(30)
```

```python
def run_for_ball():
    print("running for ball")
    red_min,blue_min=0,0
    spaces1=np.asarray([0,0,0,0,0])
    spaces2=np.asarray([0,0,0,0,0])
    for i in range(5):
        spaces1[i]=find_distance(team_red[i],ball)
    spaces1=np.sort(spaces1)
    red_min=np.argmin(spaces1)
    for i in range(5):
        spaces2[i]=find_distance(team_blue[i],ball)
    spaces2=np.sort(spaces2)
    blue_min=np.argmin(spaces2)
    a=find_distance(team_red[red_min],ball)
    b=find_distance(team_blue[blue_min],ball)
    print(a,b)
    if a<b:
        run(team_red[red_min],ball)
    elif a>b:
        run(team_blue[blue_min],ball)
    else:
        run(team_red[red_min],ball)
#     for i,j in zip(range(len(team_red)),range(len(team_blue))):
#         run(team_red[i],ball)
#         run(team_red[j],ball)
    

def confirm_posession():
    print("confirming posession")
    flag=0
    for i,j in zip(range(len(team_red)),range(len(team_blue))):
        if find_distance(ball,team_red[i])<2:
            flag=1
            print("Ball is in posession of",team_red[i].name)
            ball.x=team_red[i].x
            ball.y=team_red[i].y
        if find_distance(ball,team_blue[j])<2:
            flag=2
            print("Ball is in posession of",team_blue[j].name)
            ball.x=team_blue[j].x
            ball.y=team_blue[j].y
    update_xy()
    return flag
    
def pass_strategy():
    flags=confirm_posession()
    if flags==0:
        print("Ball Free")
    elif flags==1:
        print("Ball with Team RED")
    elif flags==2:
        print("Ball with Team BLUE")
    spaces=np.asarray([0,0,0,0,0])
    if flags==1:
        for i in range(5):
            for j in range(5):
                spaces[i]=find_distance(team_red[i],team_red[j])
        spaces=np.sort(spaces)
        spaces=np.argwhere(spaces>0)
        best=spaces[0]
#         best=np.argmax(spaces)
        temp.x=team_red[best[0]].x
        temp.y=team_red[best[0]].y
        passball(ball,team_red[best[0]])
        dribble(team_red[best[0]])
#         pass_strategy()
        start_game()
        update_xy()
    elif flags==2:
        for i in range(5):
            for j in range(5):
                spaces[i]=find_distance(team_blue[i],team_blue[j])
        spaces=np.sort(spaces)
        best=np.argmax(spaces)
        temp.x=team_blue[best].x
        temp.y=team_blue[best].y
        passball(ball,team_blue[best])
        dribble(team_blue[best])
#         pass_strategy()
        start_game()
        update_xy()
    elif flags==0:
        run_for_ball()
        pass_strategy()
```

```python
pygame.init()
gameDisplay = pygame.display.set_mode((display_width,display_height))

pygame.display.set_caption("RoboCup")

clock = pygame.time.Clock()
crashed = False
display()
start_game()
run_for_ball()

while not crashed:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            crashed = True
        
    display()
    pass_strategy()

pygame.quit()
quit()
```
